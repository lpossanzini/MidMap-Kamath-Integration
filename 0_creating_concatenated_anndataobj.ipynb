```python
import os
import glob
import anndata as ad
import scanpy as sc

# Directory containing subfolders with ..._cellbender_filtered.h5 files
base_dir = '/hpc/hers_basak/rnaseq_data/Basaklab/lpossanzini/kamath/cellbender'

# List to store individual AnnData objects
adatas = []

# Traverse the base directory to find ..._cellbender_filtered.h5 files
for root, dirs, files in os.walk(base_dir):
    for file in files:
        if file.endswith('_cellbender_filtered.h5'):
            file_path = os.path.join(root, file)
            print(f"Reading {file_path}")
            
            # Read the h5 file into an AnnData object
            adata = sc.read_10x_h5(file_path)
            
            # Ensure variable names are unique
            adata.var_names_make_unique()
            
            # Append the AnnData object to the list
            adatas.append(adata)

# Concatenate all AnnData objects into a single AnnData object
combined_adata = ad.concat(adatas, join='outer', label='batch', index_unique='-')

# Optionally, save the combined AnnData object to a file
output_file = '/hpc/hers_basak/rnaseq_data/Basaklab/lpossanzini/kamath/extra/adata_cellbender.h5ad'
combined_adata.write(output_file)
print(f"Combined AnnData object saved to {output_file}")

```
